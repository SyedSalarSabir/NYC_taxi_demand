```python
from src.paths import TRANSFORMED_DATA_DIR
import pandas as pd
from src.data_split import train_test_split
```

```python
ts_data = pd.read_parquet(TRANSFORMED_DATA_DIR/'transformed_tabular_data.parquet')
ts_data
```

```python
X_train,X_test,y_train,y_test= train_test_split(ts_data=ts_data,
                                cut_off_year=2022,
                                cut_off_month=6, cut_off_day= 1,
                                target_column='target_next_hour')
```

```python
import xgboost as xgb
```

```python
rides_col = [c for c in X_train.columns if c.startswith('rides_')]
X_train_rides_data = X_train[rides_col]
```

```python
xgb_model = xgb.XGBRegressor()

xgb_model.fit(X_train_rides_data,y_train)
```

```python
from sklearn.metrics import mean_absolute_error as mae
```

```python
X_test_rides_data = X_test[rides_col]
```

```python
xgb_predictions = xgb_model.predict(X_test_rides_data)
xgb_mae = mae(y_pred=xgb_predictions, y_true=y_test)
```
